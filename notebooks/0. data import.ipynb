```python
import os
import csv
import pandas as pd
```

```python
phd = pd.read_csv("../sources/phd.csv")
```

# Début et fin de thèse

```python
phd[["Date d'inscription première année"]] = phd[["Date d'inscription première année"]].astype(str)
phd[["Date de soutenance"]] = phd[["Date de soutenance"]].astype(str)

for idx, student in phd.iterrows():
  #print student
  bib_name = "../refs/"+str(student[2])+"/"+str(student[0]).lower()+".tsv"
  
  if student[5] != "":
    a = student[5].split("/")
    
    if len(a) > 2:
      if a[2].startswith("20"):
        a[2] = a[2][2:]
        
      phd.loc[idx, "phd year start"] = a[2]
    else:
      phd.loc[idx, "phd year start"] = None

  if student[6] != "":
    a = student[6].split("/")
    
    if len(a) > 2:
      if a[2].startswith("20"):
        a[2] = a[2][2:]

    if len(a) > 2:
      phd.loc[idx, "phd year end"] = a[2]
    else:
      phd.loc[idx, "phd year end"] = None

      
phd[["NOM", "Promotion", "phd year start", "phd year end"] ].head()
```

## add references

```python
for idx, student in phd.iterrows():
  #print student
  bib_name = "../refs/"+str(student[2])+"/"+str(student[0]).lower()+".tsv"

  before = 0
  during = 0
  after = 0
  
  # print idx
  if os.path.exists(bib_name):
    nb_citations = 0
    
    try:
      citations = pd.read_csv(bib_name, sep="\t", dtype={ "PD" : str } )
      
      nb_citations = len(citations.index)

      py = citations.groupby("PD").count()
      
      # print py
      # print student[0]

      for y, row in py.iterrows():
        phd.loc[idx, "publications "+str(y)] = row[0]
        # print y
        # print row[0]

      if phd.loc[idx, "phd year start"] != None and phd.loc[idx, "phd year end"] != None:
        #print int("20"+str(phd.loc[idx, "phd year start"]))
        for y, row in py.iterrows():
          if int(y) < int("20"+str(phd.loc[idx, "phd year start"])):
            before += int(row[0])
          elif int(y) >= int("20"+str(phd.loc[idx, "phd year start"])) and int(y) <= int("20"+str(phd.loc[idx, "phd year end"])):
            during += int(row[0])
          else:
            after += int(row[0])
            
        phd.loc[idx, "publications before"] = before
        phd.loc[idx, "publications during"] = during
        phd.loc[idx, "publications after"] = after
     
    except:
      pass
    
    phd.loc[idx, "publications"] = nb_citations   
  else:
    phd.loc[idx, "publications"] = 0


phd[["NOM", "publications", "publications during", "publications after"] ]
#phd
```

```python
phd.to_csv("../outputs/phd.csv")
```
